* now we need to split the dataset into training and testing
* trainig set is to train the algorithm
* testing set is to evaluate how good the algorith is
* a good split ratio is 66% and 33%
* the following code is defining the function

```python
import csv
import random
def loadDataset(filename, split, trainingSet=[], testSet=[]):
    with open(filename, 'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y] = dataset[x][y]
            if random.random() < split:
                trainingSet.append(dataset[x])
            else:
                testSet.append(dataset[x])
        
```

```python
trainingSet=[]
testSet=[]
loadDataset(r'iris.data.txt', 0.66, trainingSet, testSet)
print('Train:' + repr(len(trainingSet)))
print('Test:' + repr(len(testSet)))
```

* now we will create the function for the euclidian distance, this will be later used inside the get-neighbours function

```python
import math
def euclideanDistance(instance1, instance2, lenght):
    distance = 0
    for x in range(lenght):
        distance += pow((instance1[x] - instance2[x]), 2)
    return math.sqrt(distance)
```

```python
data1 = [2, 2, 2, 'a']
data2 = [4, 4, 4, 'b']
distance = euclideanDistance(data1, data2, 3)
print('Distance:' + repr(distance))
```

* now lest create the function to find nearest neighbor with a value of k

```python
import operator
def getNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```

```python
trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]
testInstance = [5, 5, 5]
k = 1
neighbors = getNeighbors(trainSet, testInstance, 1)
print(neighbors)
```

* the result above shows an example, so for testInstance = [5, 5, 5] the closest neighbor is [4, 4, 4, 'b'] and the result is that it belongs to class 'b'

* so now we will build another function that will take the output of getNeighbors as an input
* we will use this function to determine the class if you have several neighbors, much like a voting system

```python
import operator
def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
neighbors = [[1, 1, 1, 'a'], [2, 2, 2, 'a'], [3, 3, 3, 'b']]
print(getResponse(neighbors))
```

* now lets check for the accuracy
* we will pass it 2 correct 1 false

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] is predictions[x]:
            correct += 1
    return(correct/float(len(testSet))) * 100.0
```

```python
testSet = [[1, 1, 1, 'a'], [2, 2, 2, 'a'], [3, 3, 3, 'b']]
predictions = ['a', 'a', 'a']
accuracy = getAccuracy(testSet, predictions)
print(accuracy)
```
